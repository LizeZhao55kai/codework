```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

<font face="Times New Roman" size=5 color=#000000 > 
Load the data

```python
#Load the data that have been processed.
df = pd.read_csv('10m_each_processed.csv',low_memory=False)
```

<font face="Times New Roman" size=5 color=#000000 > 
Data Visualisation

```python
#Take the graph of the distribution of ages.
df['AGE'].value_counts().plot(kind='bar')
plt.title('number of Ages')
plt.xlabel('Age')
plt.ylabel('Count')
sns.despine
```

```python
#Take the graph of the gender distribution
df['SEX'].value_counts().plot(kind='bar')
plt.title('number of men and women')
plt.xlabel('Sex')
plt.ylabel('Count')
sns.despine
```

```python
#Take the graph of the employment industry distribution
df['INDSTRY'].value_counts().plot(kind='bar')
plt.title('Every Patient"s employment industry')
plt.xlabel('Industry')
plt.ylabel('Count')
sns.despine
```

```python
#Take the graph of the distribution of if the patient has work or not
df['EESTATU'].value_counts().plot(kind='bar')
plt.title('Patient Employment')
plt.xlabel('The Number of Patient Employment Categories')
plt.ylabel('Count')
sns.despine
```

```python
#Take the graph of the distribution of the quantity of services
df['QTY'].value_counts().plot(kind='bar')
plt.title('Quantity of Services')
plt.xlabel('The Quantity in Categories')
plt.ylabel('Count')
sns.despine
```

```python
#Take the graph of the distribution of the procedure categories
df['PROC1_CATEGORIES'].value_counts().plot(kind='bar')
plt.title('Procedure Categories')
plt.xlabel('Number of Every Procedure Categories')
plt.ylabel('Count')
sns.despine
```

```python
#Take the graph of the distribution of the number's diagnosis
df['DIAGNOSIS_OVERLAY'].value_counts().plot(kind='bar')
plt.title('Categories in every number"s diagnosis')
plt.xlabel('The Diagnosis Number in Categories')
plt.ylabel('Count')
sns.despine
```

```python
#Take the graph of the region distribution
df['REGION'].value_counts().plot(kind='bar')
plt.title('Region in Categories')
plt.xlabel('Region')
plt.ylabel('Count')
sns.despine
```

```python
#Take the graph of the job location distribution
df['EGEOLOC'].value_counts().plot(kind='bar')
plt.title("Patients' Job Location")
plt.xlabel('Categories of Job Location')
plt.ylabel('Count')
sns.despine
```

```python
#Take the graph to display the relationship of the therapy time and the claim 
plt.scatter(df.TIME,df.CLAIM)
plt.xlabel("Time")
plt.ylabel('Claim')
plt.title("Claim vs Time")
```

```python
#Take the graph to display the relationship of the therapy time and the claim with a joint plot(scatter and histogram).
plt.figure(figsize=(10,10))
sns.jointplot(x=df.CLAIM.values, y=df.TIME.values, height=10)
plt.ylabel('Claim', fontsize=12)
plt.xlabel('Time', fontsize=12)
plt.show()
sns.despine
```
